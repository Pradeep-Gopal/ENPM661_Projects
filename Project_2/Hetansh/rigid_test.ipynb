```python
import numpy as np
import copy
import time
import math

```

```python
def move_up(coordinates):
    coordinates[1]+=1
    return coordinates
def move_down(coordinates):
    coordinates[1]-=1
    return coordinates
def move_left(coordinates):
    coordinates[0]-=1
    return coordinates
def move_right(coordinates):
    coordinates[0]+=1
    return coordinates
def move_up_left(coordinates):
    coordinates[1]+=1
    coordinates[0]-=1
    return coordinates
def move_up_right(coordinates):
    coordinates[1]+=1
    coordinates[0]+=1
    return coordinates
def move_down_left(coordinates):
    coordinates[1]-=1
    coordinates[0]-=1
    return coordinates
def move_down_right(coordinates):
    coordinates[1]+=1
    coordinates[0]-=1
    return coordinates
def check(coordinates, radius, clearance):
    if (coordinates[0]-225)**2+(coordinates[1]-150)**2<(25+radius+clearance)**2:
        return False
    elif ((coordinates[0]-150)/(40+radius+clearance))**2+((coordinates[1]-100)/(20+radius+clearance))**2<1**2:
        return False
    elif (coordinates[0]*(-3/5)+coordinates[1]+95<0 and (coordinates[0]*(3/5)+coordinates[1]-175<0 and (coordinates[0]*(-3/5)+coordinates[1]+125>0 and (coordinates[0]*(3/5)+coordinates[1]-145>0:
        return False
    elif (coordinates[1]-185<0 and (coordinates[0]*(-7/5)+coordinates[1]-80>0 and (coordinates[0]*(-1)+coordinates[1]-100>0 and (coordinates[0]*(-13)+coordinates[1]+140<0:
        return False
    elif (coordinates[0]*(7/5)+coordinates[1]-290<0 and (coordinates[0]*(6/5)+coordinates[1]-30>0 and (coordinates[0]*(6/5)+coordinates[1]-210>0 and (coordinates[0]*(-7/5)+coordinates[1]-80<0:
        return False
    elif (coordinates[0]*(-9/5)+coordinates[1]+144>0 and (coordinates[0]*(7/65)+coordinates[1]-(523/13)>0 and (coordinates[0]*(-9/5)+coordinates[1]+17<0 and (coordinates[0]*(7/65)+coordinates[1]-(-647/13)<0:
        return False
    elif coordinates[0]>300 or coordinates[0]<0 or coordinates[1]>200 or coordinates[1]<0:
        return False
    else:
        return True
def check_unique(history, position):
    
    if position in history:
        return False
    else:
        return True
```

```python
def main():
    
    
    while True:
        radius = int(input("enter radius of the robot "))
        if radius>0:
            break
        else:
            print("Radius must be positive. Please enter again")
    
    while True:
        clearance = int(input("enter clearance for the robot "))
        if clearance>0:
            break
        else:
            print("clearance must be positive. Please enter again")
            
    
    
    while True:
    
        print("Enter the coordinates of starting point : ")
        starting_node = list(map(int, input().split())) 
        if check(starting_node, radius, clearance)==True:
            break
            
        else:
            print("The starting node is invalid. Enter again.")
    
    
    while True:
        
        print("Enter the coordinates of target point : ")
        target_node = list(map(int, input().split())) 
        if check(target_node, radius, clearance)==True:
            break
        else:
            print("The target node is invalid. Enter again.")
        
    
    
    
    import time

    start = time.time()
    
    


    current_node= []  #List to store all the values of nodes
    parent= [] #List to store the index value of parent node
    current_node.append(starting_node)#first value of this list will be the initial matrix and final value will be target matrix
    parent.append(0)
    x=0
    
    while True:
        
        
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_up(temp))
        if check(temp, radius, clearance)==True:
            if check_unique(current_node,temp)==True:
                current_node.append(temp)
                parent.append(x)
                print(temp)
                
                
                if np.array_equal(temp, target_node):
                    print("Matrix found")
                    break
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_down(temp))
        if check(temp, radius, clearance)==True:
            if check_unique(current_node,temp)==True:
                current_node.append(temp)
                parent.append(x)
                print(temp)
                
                if np.array_equal(temp, target_node):
                    print("Matrix found")
                    break
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_left(temp))
        if check(temp, radius, clearance)==True:
            if check_unique(current_node,temp)==True:
                current_node.append(temp)
                parent.append(x)
                print(temp)
                
                if np.array_equal(temp, target_node):
                    print("Matrix found")
                    break
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_right(temp))
        if check(temp, radius, clearance)==True:
            if check_unique(current_node,temp)==True:
                current_node.append(temp)
                parent.append(x)
                print(temp)
                
                if np.array_equal(temp, target_node):
                    print("Matrix found")
                    break
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_up_left(temp))
        if check(temp, radius, clearance)==True:
            if check_unique(current_node,temp)==True:
                current_node.append(temp)
                parent.append(x)
                print(temp)
                
                if np.array_equal(temp, target_node):
                    print("Matrix found")
                    break
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_up_right(temp))
        if check(temp, radius, clearance)==True:
            if check_unique(current_node,temp)==True:
                current_node.append(temp)
                parent.append(x)
                print(temp)
                if np.array_equal(temp, target_node):
                    print("Matrix found")
                    break
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_down_left(temp))
        if check(temp, radius, clearance)==True:
            if check_unique(current_node,temp)==True:
                current_node.append(temp)
                parent.append(x)
                print(temp)
                if np.array_equal(temp, target_node):
                    print("Matrix found")
                    break
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_down_right(temp))
        if check(temp, radius, clearance)==True:
            if check_unique(current_node,temp)==True:
                current_node.append(temp)
                parent.append(x)
                print(temp)
                if np.array_equal(temp, target_node):
                    print("Matrix found")
                    break
        
        
        x+=1
    
    end = time.time()
    print("Time taken:",end-start)
        
    back_path = []
    length = len(parent) - 1
    back_path.append(current_node[length])
    while not(length==0):
        temp_index=copy.deepcopy(parent[length])
        back_path.append(current_node[temp_index])
        length=temp_index
   
    back_path.reverse()
    
    #print(current_node)
    print(back_path)
    
if __name__ == '__main__':
    main()
    
```
