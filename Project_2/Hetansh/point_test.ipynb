```python
import numpy as np
import copy
import time
import math
import matplotlib.pyplot as plt
```

```python
def move_up(coordinates):
    coordinates[1]+=1
    return coordinates
def move_down(coordinates):
    coordinates[1]-=1
    return coordinates
def move_left(coordinates):
    coordinates[0]-=1
    return coordinates
def move_right(coordinates):
    coordinates[0]+=1
    return coordinates
def move_up_left(coordinates):
    coordinates[1]+=1
    coordinates[0]-=1
    return coordinates
def move_up_right(coordinates):
    coordinates[1]+=1
    coordinates[0]+=1
    return coordinates
def move_down_left(coordinates):
    coordinates[1]-=1
    coordinates[0]-=1
    return coordinates
def move_down_right(coordinates):
    coordinates[1]+=1
    coordinates[0]-=1
    return coordinates
```

```python
def main():
    
    coordinates = np.zeros([200, 300])
    for i in range(200):
        for j in range(300):
        
            if (j-225)**2+(i-150)**2<25**2:
                coordinates[i][j] = 0
            elif ((j-150)/40)**2+((i-100)/20)**2<1**2:
                coordinates[i][j] = 0
            elif (j*(-3/5)+i+95)<0 and (j*(3/5)+i-175)<0 and (j*(-3/5)+i+125)>0 and (j*(3/5)+i-145)>0:
                coordinates[i][j] = 0
            elif i-185<0 and j*(-7/5)+i-80>0 and j*(-1)+i-100>0 and j*(-13)+i+140<0:
                coordinates[i][j] = 0
            elif j*(7/5)+i-290<0 and j*(-6/5)+i-30>0 and j*(6/5)+i-210>0 and j*(-7/5)+i-80<0:
                coordinates[i][j] = 0
            elif i - (1.73)*j + 135 > 0 and i + (0.58)*j - 96.35  <= 0 and i - (1.73)*j - 15.54 <= 0 and i + (0.58)*j - 84.81 >= 0:
                coordinates[i][j] = 0
            elif j>=299 or j<=0 or i>=199 or i<=0:
                coordinates[i][j] = 0
            else:
                coordinates[i][j] = 1
    
    
    while True:
    
        print("Enter the coordinates of starting point : ")
        starting_node = list(map(int, input().split())) 
        if coordinates[starting_node[1]][starting_node[0]]==1:
            
            break
            
        else:
            print("The starting node is invalid. Enter again.")
    
    
    while True:
        
        print("Enter the coordinates of starting point : ")
        target_node = list(map(int, input().split())) 
        if coordinates[target_node[1]][target_node[0]]==1:
            coordinates[target_node[1]][target_node[0]]=5
            break
            
        else:
            print("The starting node is invalid. Enter again.")
    
    
    import time

    start = time.time()
    
    


    current_node= []  #List to store all the values of nodes
    parent= [] #List to store the index value of parent node
    current_node.append(starting_node)#first value of this list will be the initial matrix and final value will be target matrix
    parent.append(0)
    x=0
    
    while True:
        
        
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_up(temp))
        if coordinates[temp[1]][temp[0]]==5:
            current_node.append(temp)
            parent.append(x)
            print("Matrix found")
            break
        elif coordinates[temp[1]][temp[0]]==1:
            current_node.append(temp)
            parent.append(x)
            
            coordinates[temp[1]][temp[0]]=0
                
                
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_down(temp))
        if coordinates[temp[1]][temp[0]]==5:
            current_node.append(temp)
            parent.append(x)
            print("Matrix found")
            break
        elif coordinates[temp[1]][temp[0]]==1:
            current_node.append(temp)
            parent.append(x)
            
            coordinates[temp[1]][temp[0]]=0
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_left(temp))
        if coordinates[temp[1]][temp[0]]==5:
            current_node.append(temp)
            parent.append(x)
            print("Matrix found")
            break
        elif coordinates[temp[1]][temp[0]]==1:
            current_node.append(temp)
            parent.append(x)
            
            coordinates[temp[1]][temp[0]]=0
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_right(temp))
        if coordinates[temp[1]][temp[0]]==5:
            current_node.append(temp)
            parent.append(x)
            print("Matrix found")
            break
        elif coordinates[temp[1]][temp[0]]==1:
            current_node.append(temp)
            parent.append(x)
            
            coordinates[temp[1]][temp[0]]=0
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_up_right(temp))
        if coordinates[temp[1]][temp[0]]==5:
            current_node.append(temp)
            parent.append(x)
            print("Matrix found")
            break
        elif coordinates[temp[1]][temp[0]]==1:
            current_node.append(temp)
            parent.append(x)
            
            coordinates[temp[1]][temp[0]]=0
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_up_left(temp))
        if coordinates[temp[1]][temp[0]]==5:
            current_node.append(temp)
            parent.append(x)
            print("Matrix found")
            break
        elif coordinates[temp[1]][temp[0]]==1:
            current_node.append(temp)
            parent.append(x)
            
            coordinates[temp[1]][temp[0]]=0
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_down_right(temp))
        if coordinates[temp[1]][temp[0]]==5:
            current_node.append(temp)
            parent.append(x)
            print("Matrix found")
            break
        elif coordinates[temp[1]][temp[0]]==1:
            current_node.append(temp)
            parent.append(x)
            
            coordinates[temp[1]][temp[0]]=0
        
        temp= copy.deepcopy(current_node[x])
        temp = copy.deepcopy(move_down_left(temp))
        if coordinates[temp[1]][temp[0]]==5:
            current_node.append(temp)
            parent.append(x)
            print("Matrix found")
            break
        elif coordinates[temp[1]][temp[0]]==1:
            current_node.append(temp)
            parent.append(x)
            
            coordinates[temp[1]][temp[0]]=0
        
        
        x+=1
        
    back_path = []
    length = len(parent) - 1
    back_path.append(current_node[length])
    while not(length==0):
        temp_index=copy.deepcopy(parent[length])
        back_path.append(current_node[temp_index])
        length=temp_index
   
    back_path.reverse()
    print(back_path)
    
    coo = np.zeros([200, 300])
    for i in range(200):
        for j in range(300):
        
            if (j-225)**2+(i-150)**2<25**2:
                coo[199-i][j] = 0
            elif ((j-150)/40)**2+((i-100)/20)**2<1**2:
                coo[199-i][j] = 0
            elif (j*(-3/5)+i+95)<0 and (j*(3/5)+i-175)<0 and (j*(-3/5)+i+125)>0 and (j*(3/5)+i-145)>0:
                coo[199-i][j] = 0
            elif i-185<0 and j*(-7/5)+i-80>0 and j*(-1)+i-100>0 and j*(-13)+i+140<0:
                coo[199-i][j] = 0
            elif j*(7/5)+i-290<0 and j*(-6/5)+i-30>0 and j*(6/5)+i-210>0 and j*(-7/5)+i-80<0:
                coo[199-i][j] = 0
            elif i - (1.73)*j + 135 > 0 and i + (0.58)*j - 96.35  <= 0 and i - (1.73)*j - 15.54 <= 0 and i + (0.58)*j - 84.81 >= 0:
                coo[199-i][j] = 0
            elif j>300 or j<0 or i>200 or i<0:
                coo[199-i][j] = 0
            else:
                coo[199-i][j] = 1
    for i in range(len(back_path)):
        coo[199-(back_path[i][1])][(back_path[i][0])] = 5
    
    

    
    plt.matshow(coo)
    end = time.time()
    print("Time taken:",end-start)

    plt.show()
    
if __name__ == '__main__':
    main()
```
